인덱스를 0부터 시작하는 숫자가 아닌 다른 값으로 변경 가능

```python
df.index = df["account"]    # account 열을 인덱스로 변경
df.head()
```

```python
df.index = df["account"]    # account 열을 인덱스로 변경
del df["account"]           # account 열 삭제
df.head()
```

loc 함수  인덱스 이름과 열 이름으로 데이터 추출

```python
df.loc[[211829,320563],["name","street"]]       # 특정 account 번호에 해당하는 데이터 추출이 용이해짐  loc[열, 행] 열로 211829,320563, 행으로 name, street
```

```python
df.loc[205217:,["name","street"]]   # 인덱스 값은 정렬되지 않고 저장된 순서대로 출력        #열 205217 이후 부터 모든 값. 행 name, street
```

.iloc() 인덱스 위치로만 데이터 추출 0부터 시작.  


```python
df.iloc[:10, :3]        # 열을 기준으로 10개, 행을 기준으로 3개까지
```

reset_index 새로운 인덱스가 할당된 객체 생성.  지정했던 인덱스 값들을 다시 리셋함. 

```python
df_new = df.reset_index()
df_new
```

.drop() 특정 열이나 행을 삭제한 객체를 반환 - 해당 값이 실제로 사라지지 않음  

```python
df_new.drop(1).head()   # 인덱스 값이 1인 행 삭제(단 실제로 삭제 되지는 않음)
```

```python
df_new.drop(0).head()   # 여기서는 0번을 삭제   
```

```python
df_new.head()   # 실제로 삭제가 되지는 않음
```

```python
df_drop = df_new.drop(1)    #실제로 삭제된 값을 기억하려면 새롭게 객체에다 지정
df_drop
```

원본 객체에 영향을 주고 싶다면 inplace 매개변수를 사용 

```python
df_new.drop(1, inplace=True)        # 값이 반영됨.  행 1번이 사라지고, 그 값이 반영 됨.
df_new
```

특정 열이나 행을 삭제 가능.

```python
df_new.drop("account", axis=1).head() #account열 제거
```

```python
df_new.drop(["account", "name"], axis=1).head() #account, name열 제거       axis 1 = 행
```
