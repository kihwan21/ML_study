where() 배열이 boolean으로 이루어져 있을 때 참인 값들의 인덱스를 반환 

```python
x = np.array([4, 6, 7, 3, 2])
x > 5      
```

```python
np.where(x>5)  # x > 5를 만족하는 값은 6과 7  6과 7의 인덱스 값인 [1,2] 반환
```

```python
x = np.array([4, 6, 7, 3, 2])
np.where(x>5 , 10, 20)      # true면 10을 출력 false면 20을 출력
```

```python
x = np.array([4, 6, 7, 3, 2])
np.where(x>5 , 0, 1) # true면 0 false면 1 출력
```

boolean index 배열에 있는 값들을 반환할 특정 조건을 불린형의 배열에 넣어서 추출  
인덱스에 들어가는 배열은 불린형 이어야함.  
불린형 배열과 추출대상이 되는 배열의 구조가 같아야함. 

```python
x = np.array([4, 6, 7, 3, 2])
x > 3       #true인 값만 출력
```

```python
cond = x > 3
x[cond] #x가 3이상인 것을 변수 cond에 넣고(마스킹)하고 x를 출력함.
```

```python
x.shape #배열의 구조가 같음.
```

```python
cond.shape  # 배열의 구조가 같음
```

fancy index - 정수형 배열의 값을 사용하여 해당 정수의 인덱스에 위치한 값을 반환  

데이터의 인덱스를 마스킹 함. 

```python
x = np.array([4, 6, 7, 3, 2])
cond = np.array([1, 2, 0, 2, 2, 2,], int)       # 인덱스를 마킹함.
x[cond]     # x에 있는 데이터가 마스킹을 한 값에 따라서 인덱스가 이동함.
```

```python
x.take(cond) # x에 있는 데이터가 마스킹을 한 값에 따라서 인덱스가 이동함.

cond_1 = np.array([1,2,0,4,0,0])
x.take(cond_1)
```
