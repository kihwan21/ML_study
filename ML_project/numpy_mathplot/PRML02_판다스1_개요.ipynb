# 판다스(Pandas) - 개요

판다스 파이썬의 데이터 분석 라이브러리  
데이터 테이블을 다루는 도구  
  
DataFrame: 데이터 테이블 전체 객체  (matrix)
Series 각 열 데이터를 다루는 객체 (colum) 

시리즈 객체: 피쳐벡터와 같은 개념.  (각 열 데이터를 다룸)
1차원데이터를 저장하기 위한 데이터 구조  
생성된 데이터프레임 안에 포함될 수 있음.  
list, dict, ndarray등 다양한 데이터 타입이 시리즈 객체 형태로 변환되기도 함.   

각 데이터를 구분해 주는것이 주요 역할  

```python
import pandas as pd #pandas 모듈 호출
import numpy as np  #numpy 모듈 호출

from pandas import Series, DataFrame

list_data = [1,2,3,4,5]
list_name = ["a","b","c","d","e"]
example_obj = Series(data = list_data, index=list_name) # 데이터 값 지정    인덱스 값 지정
example_obj
```

index label 지정 가능.  
중복도 허용 즉 다양한 인덱스 레이블을 지정 가능함.  

데이터: 기존 다른 객체처럼 값을 저장하는 요소  
인덱스, 항상 0 부터 시작하고 숫자로만 할당  
데이터 타입: 넘파이의 데이터 타입과 일치   
  
  문자열   object  
  정수     int64  
  실수     float64  
  참 거짓  bool  
  날짜     datetime64

```python
dat2 = pd.Series(np.array([1,3,5,7]))
dat3 = pd.Series(np.array(['a', 'b', 'c', 'd']))
dat4 = pd.Series(np.array([1,'a',3.14,'b']))
print(dat2)
print(dat3)
print(dat4)  

# 다양한 데이터 타입을 Series 안에 넣을 수 있음. 
```

시리즈 객체 생성하기  
데이터프레임 객체를 먼저 생성하고 각 열에서 시리즈 객체를 뽑는 것이 일반적인 방법  
다양한 시퀀스형 데이터 타입으로 저장가능 

```python
dict_data = {"a":1, "b":2, "c":3, "d":4, "e":5}
example_obj = Series(dict_data, dtype=np.float32, name="example_data")  # 시리즈에 타입을 넣고, 이름을 'example_data'로 지정함
example_obj
```

만든 데이터 값보다 많은 인덱스 값을 추가하면 NaN값이 출력됨.

### 데이터프레임 객체  
데이터 테이블 전체를 지칭하는 객체  (matrix)
넘파이 배열의 특성을 그대로 가짐  
  
인덱싱: 열과 행 각각 사용하여 하나의 데이터에 접근 

.read_확장자명() 데이터 로딩(데이터 읽기)

```python
#데이터 URL을 변수 data_url에 넣기                                                                  # uci 검색해보기 
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data' 
#csv 데이터 로드
df_data = pd.read_csv(data_url, sep='\s+' , header = None)  # sep='\s+' 구분자
df = pd.DataFrame(df_data)  #읽어온 데이터를 데이터프레임 형태로 저장
df
```

```python
from pandas import Series, DataFrame
# dict 타입 데이터에서 key는 열 이름, value(값)은 시퀀스형 데이터 타입을 넣어 각 열의 데이터로 만듦.
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'age': [42, 52, 36, 24, 73],
            'city': ['San Francisco', 'Baltimore', 'Miami', 'Douglas', 'Boston']}
df = DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'city'])      # 데이터 프레임 객체를 생성하기 위해 DataFrame 함수 호출. 열은 지정해도 되고 안해도 됨.
df
```
