### 3차원 텐서에서는 어케 할까?  
  
3차원에서는   
axis = 0은 차원   
axis = 1은 열  
axis = 2는 행  


```python
test_array = np.arange(1, 13).reshape(3, 4)
third_order_tensor = np.array([test_array, test_array, test_array])
third_order_tensor
```

```python
third_order_tensor.sum(axis=0)  # 차원마다 더하기 
```

```python
third_order_tensor.sum(axis=1)  # 열 마다 더하기 
```

```python
third_order_tensor.sum(axis=2)  # 행 마다 더하기    
```

연산함수  mean(평균), std(표준편차 연산), sqrt(제곱근 연산)  
  
연산 함수:배열 내부 연산을 지원하는 함수

```python
test_array = np.arange(1, 13).reshape(3, 4) 
test_array
```

```python
test_array.mean(axis=1) #axis=1  축을(행) 기준으로 평균 연산. mean 평균 
```

```python
test_array.std() #전체 값에 대한 표준편차 연산  
```

```python
test_array.std(axis=0) #axis=0 축을(열) 기준으로 표준편차 연산
```

```python
np.sqrt(test_array) #각 요소에 제곱근 연산 수행
```

### 연결함수  
vstack 수직으로 붙여 하나의 행렬 생성  
hstack 수평으로 하나의 행렬 생성

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
np.vstack((v1, v2)) # 수직으로 붙여서 하나의 행렬 생성
```

```python
np.hstack((v1,v2))  # 수평으로 붙여서 하나의 행렬 생성
```

```python
v1 = v1.reshape(-1, 1) #v1 = np.array([1, 2, 3])
v2 = v2.reshape(-1, 1) #v2 = np.array([4, 5, 6])
v1
```

```python
v2
```

```python
np.hstack((v1,v2))  #v1, v2를 수평으로 행렬을 만듦. [1][4] 는 각각의 행렬에서 동일한 위치이기 때문에 [1,4]가 수평으로 붙음.  다른 행렬도 마찬가지
```

```python
np.vstack((v1,v2)) #v1, v2를 수직으로 행렬을 만듦. [1][4] 는 각각의 행렬에서 동일한 위치이기 때문에 [1], [2]가 수직으로 붙음.  다른 행렬도 마찬가지
```

concatenate 함수 축을 고려하여 두 개의 배열을 결합  
앞선 스택 계열의 함수와는 달리 생성될 배열과 소스가 되는 배열의 축방향 차원이 같아야함.   
두벡터를 결합하고 싶다면, 해당 벡터를 일단 2차원 배열 꼴로 변환 후 행렬로 나타내야 함.

```python
v1 = np.array([[1, 2, 3]])
v2 = np.array([[4, 5, 6]])
v3 = np.concatenate((v1,v2), axis=0) # 열 방향
v3
```

```python
v1 = np.array([[1, 2, 3]])
v2 = np.array([[4, 5, 6]])
np.concatenate((v1,v2), axis=1) # 행 방향
```

```python
v1 = np.array([1, 2, 3, 4]).reshape(2,2)
v1
```

```python
v2 = np.array([[5,6]]).T    # .T 전치 / np.transpose(대상_변수)   / np.swqpaxws(변수, axis(0 or 1), axis(0 or 1))
v2
```

```python
np.concatenate((v1,v2), axis=1) #axis 1은 행 / 0은 열
```

```python
np.concatenate((v1,v2), axis=0) #축 방향 차원이 달라서 안됨. v1이 축 방향에 더 많아서 안됨
```

### 넘파이 사칙연산  
같은 배열의 구조일 때 요소별 연산이 가능하다.

```python
x = np.arange(1, 7).reshape(2,3)
x
```

```python
x + x
```

```python
x - x
```

```python
x / x
```

```python
x ** x
```

배열 간의 곱셈 m*n n*l 행렬 내적하면 m*l 행렬 생성  
행렬의 곱셈 연산 .dot()

```python
x_1 = np.arange(1, 7).reshape(2,3)
x_1
```

```python
x_2 = np.arange(1, 7).reshape(3,2)
x_2
```

```python
x_1.dot(x_2)    #dot 벡터의 곱셈 연산 가능
```
