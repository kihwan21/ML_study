# 판다스(Pandas) - 그룹별 집계

그룹별 집계  
동일한 객체를 가진 데이터만 뽑아 기술통계 데이터를 추출  



```python
import pandas as pd #pandas 모듈 호출
import numpy as np  #numpy 모듈 호출

ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings', 'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
            'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
            'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
            'Points':[876,789,863,673,741,812,756,788,694,701,804,690]
            }

df = pd.DataFrame(ipl_data)
df
```

집계 요약된 통계정보 추출  
agg함수:  min, 넘파이 mean등 기존함수 그대로 적용

```python
df.groupby("Team")["Points"].sum()      # 그룹핑의 기준 열() ('team') ['Points']적용 받는 열 .sum() 적용받는 연산 
```

멀티 인덱스 그룹별 집계 - 한 개 이상의 열을 기준으로 그룹별 집계 실행     - 해당 하는 열을 기준으로 그룹별 집계가 실행됨.

리스트를 사용하여 여러개의 열 이름을 기준으로 넣으면 여러 열이 키 값이 되어 출력   
계층적 인덱스 형태(hierarchical index) 

한 개 이상의 열로 그룹별 집계를 수행하면 여러 열이 모두 인덱스로 반환됨.  - 그룹별로 묶어서 한번에 연산하기 쉽게 해줌.
groupby("열")["열"].연산자()

```python
mlt_group = df.groupby("Team")["Rank"].sum()
mlt_group
```

```python
multi_groupby = df.groupby(["Team", "Year"])["Points"].sum()        # 기준 ([열,열])    [열]적용 받는 열  
multi_groupby
```

```python
multi_groupby.index # 여러 열이 모두 인덱스로 반환됨. 앞서 그룹화 했던 모든 열들이 인덱스로 변환이 됨.
```

한 개 이상의 열로 그룹별 집계를 수행하면 여러 열이 모두 인덱스로 반환됨.

```python
multi_groupby["Devils":"Kings"]     # 그룹화 한 것 중에서 열이 "Devils"에서 부터 "Kings"에 해당하는 값을 모두 가져와라
```

```python
multi_groupby["Devils":"Riders"]    # 그룹화 한 것 중에서 열이 "Devils"에서 부터 "Riders"에 해당하는 값을 모두 가져와라
```

```python
multi_groupby
```

두번째 인덱스를    열로 변화시켜 행렬의 형태로 데이터를 보여줌  

```python
multi_groupby.unstack()     # 두번쨰 인덱스 값인 "Year"가 위치를 옮겨서 행렬의 형태로 바뀜. <---- 위에 데이터랑 참고
```

.swaplevel()    : 인덱스 간 레벨을 변경  
.sort_index()   : 첫 번째 인덱스를 기준으로 데이터 재정렬

```python
multi_groupby.swaplevel()   # Team 과 Year의 인덱스 값이 서로 바뀜.
```

```python
multi_groupby.swaplevel().sort_index()      # 첫 번째 인덱스 값인 2014를 기준으로 값이 정렬됨.
```

그룹화된 상태: 분할 -> 적용 -> 결합 중에서 분할까지만 이루어진 상태

.get_group() 해당 키 값을 기준으로 분할된 데이터프레임 객체를 확인 

```python
grouped = df.groupby("Team")        # grouped 안에 team으로 그룹화 된 상태 
grouped.get_group("Riders")     # team으로 그룹화된 상태인데 .get_group()으로 Riders만 다시 가져옴
```

집계 agg() 요약된 통계 정보 추출   
agg 함수들 min, 넘파이, mean 등 기존 함수 그대로 적용 가능

```python
grouped.agg(min)        # 모든 칼럼 마다 각 함수를 적용 시켜줌. 
```

```python
grouped.agg(sum)
```

```python
grouped.agg(max)
```

```python
grouped.agg(np.mean)
```

변환(transformation) : 해당 정보를 변환  
키 값 별로 요약된 정보가 아닌 개별 데이터 변환 지원 


```python
grouped.transform(max)      # 데이터를 바꿈. # 이미 grouped 에 team으로 그룹화가 된 상태.  grouped안에 있는 열 값들을 max로 출력 함.    햇갈리면 개발환경 5 110 페이지 그림 보기
```

이미 있는 함수가 아닌 새롭게 정의 한 값도 적용이 가능함. 

```python
score = lambda x : (x - x.mean()) / x.std()  #def 도 적용이 됨. 인자를 마음대로 정의 가능. # 여기서는 표준편차 공식을 사용해서 표준편차를 구하는 함수를 만듦
grouped.transform(score)
```

필터링도 가능. 여기서는 필터 안에 람다 함수를 넣음 

```python
df.groupby('Team').filter(      # 데이터가 팀별로 3개 이상인 것 
    lambda x: len(x) >= 3)
```
