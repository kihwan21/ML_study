# 인터넷 코드

```python
#사전준비
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
#데이터 읽기
df = sns.load_dataset('titanic')
df.head()
```

```python
#결손값 확인
df.isnull().sum()
```

```python
# 나이만 결손값 처리
age_mean = df['age'].mean()
df['age'] = df['age'].fillna(age_mean) # 나이를 모르는 데이터들 때문에 값을 넣기 위해 평균으로 넣음 
```

```python
# 목적 변수와 설명 변수의 분할
df_x = df.drop(['survived','alive'], axis=1)
df_y = df['survived']
```

```python
df_x
```

```python
# 목적 변수의 더미 변수화
df_x = pd.get_dummies(df_x)
df_x
```

```python
# 학습용-테스트용 데이터로 분할
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df_x,df_y,random_state=1)
```

```python
# 랜덤 포레스트 모델 생성
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=200,random_state=1)
```

```python
model.fit(train_x, train_y)
```

```python
# 랜덤 포레스트 모델의 점수
print(model.score(test_x,test_y))
```

# 어레인지 코드

```python
#사전준비
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
#데이터 읽기
df = sns.load_dataset('titanic')
df.head()
```

```python
#결손값 확인
df.isnull().sum()
```

```python
# 나이를 평균값으로 대체
df['age'] = df['age'].fillna(df['age'].mean())
df.isnull().sum()
```

```python
# 성별 예측
df_x = df.drop(['sex', 'who', 'adult_male'], axis=1)
df_y = df['sex']
```

```python
# 생존 예측
df_x = df.drop(['survived','alive'], axis=1)
df_y = df['survived']
```

```python
# 목적 변수의 더미 변수화
df_x = pd.get_dummies(df_x)
```

```python
# 학습용-테스트용 데이터로 분할
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df_x,df_y,random_state=2)
```

```python
# 랜덤 포레스트 모델 생성
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=150,random_state=2)
```

```python
model.fit(train_x, train_y)
```

```python
# 랜덤 포레스트 모델의 점수
print(model.score(test_x,test_y))
```

# 테스트

```python
from sklearn.ensemble import RandomForestClassifier
model_randomforest = RandomForestClassifier(n_estimators=100,random_state=1)
model_randomforest.fit(train_x,train_y)

from sklearn import tree
model_tree = tree.DecisionTreeClassifier(max_depth=5, random_state=1)
model_tree.fit(train_x, train_y)

print('랜덤포래스트:',model_randomforest.score(test_x,test_y))
print('의사결정트리:',model_tree.score(test_x,test_y))
```
