```python
# Lasso

import mglearn

import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.linear_model import Lasso



X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)



lasso = Lasso().fit(X_train, y_train)
print()

print("훈련 세트의 정확도 : {:.2f}".format(lasso.score(X_train, y_train)))

print("테스트 세트의 정확도 : {:.2f}".format(lasso.score(X_test, y_test)))

print("사용한 특성의 수 : {}".format(np.sum(lasso.coef_ != 0)))

print("사용한 max_iter : {}".format(lasso.n_iter_))



lasso001 = Lasso(alpha=0.001, max_iter=100000).fit(X_train, y_train)
print()

print("훈련 세트의 정확도 : {:.2f}".format(lasso001.score(X_train, y_train)))

print("테스트 세트의 정확도 : {:.2f}".format(lasso001.score(X_test, y_test)))

print("사용한 특성의 수 : {}".format(np.sum(lasso001.coef_ != 0)))

print("사용한 max_iter : {}".format(lasso001.n_iter_))



lasso0001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print()

print("훈련 세트의 정확도 : {:.2f}".format(lasso0001.score(X_train, y_train)))

print("테스트 세트의 정확도 : {:.2f}".format(lasso0001.score(X_test, y_test)))

print("사용한 특성의 수 : {}".format(np.sum(lasso0001.coef_ != 0)))

print("사용한 max_iter : {}".format(lasso0001.n_iter_))

```

```python
# Ridge

import mglearn

import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.linear_model import Ridge


X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)



ridge = Ridge().fit(X_train, y_train)
print()

print("훈련 세트의 정확도 : {:.2f}".format(ridge.score(X_train, y_train)))

print("테스트 세트의 정확도 : {:.2f}".format(ridge.score(X_test, y_test)))

print("사용한 특성의 수 : {}".format(np.sum(ridge.coef_ != 0)))



ridge0 = Ridge(alpha=0.1).fit(X_train, y_train)
print()

print("훈련 세트의 정확도 : {:.2f}".format(ridge0.score(X_train, y_train)))

print("테스트 세트의 정확도 : {:.2f}".format(ridge0.score(X_test, y_test)))

print("사용한 특성의 수 : {}".format(np.sum(ridge0.coef_ != 0)))



ridge000 = Ridge(alpha=0.0001).fit(X_train, y_train)
print()

print("훈련 세트의 정확도 : {:.2f}".format(ridge000.score(X_train, y_train)))

print("테스트 세트의 정확도 : {:.2f}".format(ridge000.score(X_test, y_test)))

print("사용한 특성의 수 : {}".format(np.sum(ridge000.coef_ != 0)))



```

```python
# ElasticNet

import mglearn

import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.linear_model import ElasticNet



X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)



elasticnet = ElasticNet().fit(X_train, y_train)
print()

print("훈련 세트의 정확도 : {:.2f}".format(elasticnet.score(X_train, y_train)))

print("테스트 세트의 정확도 : {:.2f}".format(elasticnet.score(X_test, y_test)))

print("사용한 특성의 수 : {}".format(np.sum(elasticnet.coef_ != 0)))

print("사용한 max_iter : {}".format(elasticnet.n_iter_))



elasticnet00 = ElasticNet(alpha=0.001, max_iter=100000).fit(X_train, y_train)
print()

print("훈련 세트의 정확도 : {:.2f}".format(elasticnet00.score(X_train, y_train)))

print("테스트 세트의 정확도 : {:.2f}".format(elasticnet00.score(X_test, y_test)))

print("사용한 특성의 수 : {}".format(np.sum(elasticnet00.coef_ != 0)))

print("사용한 max_iter : {}".format(elasticnet00.n_iter_))



elasticnet000 = ElasticNet(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print()

print("훈련 세트의 정확도 : {:.2f}".format(elasticnet000.score(X_train, y_train)))

print("테스트 세트의 정확도 : {:.2f}".format(elasticnet000.score(X_test, y_test)))

print("사용한 특성의 수 : {}".format(np.sum(elasticnet000.coef_ != 0)))

print("사용한 max_iter : {}".format(elasticnet000.n_iter_))


```
