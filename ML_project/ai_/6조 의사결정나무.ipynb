# 인터넷 코드

```python
#사전 준비
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
df = sns.load_dataset('titanic')
df.head()
```

```python
#결정트리를 가시화하기 위해 성별, 티켓, 운임만
df_x = df[['sex','pclass','fare']]
df_y = df['survived']
```

```python
#문자행을 숫차로 변환
df_x = pd.get_dummies(df_x, drop_first=True)
```

```python
#train_test_split를 사용해서 학습용과 테스트용으로 분할한다.
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df_x,df_y,random_state=1)
```

```python
#결정트리의 모델을 작성
from sklearn import tree

#max_depth는 트리의 최대 깊이를 표현하고 조건을 얼마나 분기시킬까 설정
model = tree.DecisionTreeClassifier(max_depth=2, random_state=1)
```

```python
#fit로 학습을 시킨다.
model.fit(train_x, train_y)
```

```python
#predict로 예측을 실치
model.predict(test_x)
```

```python
#score로 정답율을 낸다.
model.score(test_x,test_y)
```

```python
#plot_tree로 결정트리 모델을 트리 구조로 가시화
from sklearn.tree import plot_tree
plot_tree(model, feature_names=train_x.columns, class_names=True, filled=True)
```

## 특성을 바꿔서 진행

```python
# 사전 준비
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
df = sns.load_dataset('titanic')
df.head()
```

```python
# 선택한 특성
df_x = df[['sex','fare', 'adult_male', 'alone', 'pclass']]
df_y = df['survived']
```

```python
#문자행을 숫차로 변환
df_x = pd.get_dummies(df_x, drop_first=True)

#train_test_split를 사용해서 학습용과 테스트용으로 분할한다.
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df_x,df_y,random_state=1)

#결정트리의 모델을 작성
from sklearn import tree

#max_depth는 트리의 최대 깊이를 표현하고 조건을 얼마나 분기시킬까 설정
model = tree.DecisionTreeClassifier(max_depth=7, random_state=1)

#fit로 학습을 시킨다.
model.fit(train_x, train_y)

```

```python
#predict로 예측을 실치
model.predict(test_x)
```

```python
#score로 정답율을 낸다.
model.score(test_x,test_y)
```

```python
#plot_tree로 결정트리 모델을 트리 구조로 가시화
from sklearn.tree import plot_tree
plot_tree(model, feature_names=train_x.columns, class_names=True, filled=True)
```

```python
# 선택한 특성
df_x = df[['survived','fare', 'alone', 'pclass']]
df_y = df['sex']
```

```python
#문자행을 숫차로 변환
df_x = pd.get_dummies(df_x, drop_first=True)

#train_test_split를 사용해서 학습용과 테스트용으로 분할한다.
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df_x,df_y,random_state=1)

#결정트리의 모델을 작성
from sklearn import tree

#max_depth는 트리의 최대 깊이를 표현하고 조건을 얼마나 분기시킬까 설정
model = tree.DecisionTreeClassifier(max_depth=5, random_state=1)

#fit로 학습을 시킨다.
model.fit(train_x, train_y)

```

```python
#predict로 예측을 실치
model.predict(test_x)
```

```python
#score로 정답율을 낸다.
model.score(test_x,test_y)
```
