```python
import pandas as pd

#데이터
xy_data = pd.read_csv("churn-train.csv")
X_test = pd.read_csv("churn-test.csv")

#데이터 전처리

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
xy_data.gender = le.fit_transform(xy_data.gender) 
xy_data.Partner = le.fit_transform(xy_data.Partner)
xy_data.Dependents = le.fit_transform(xy_data.Dependents)
xy_data.PhoneService = le.fit_transform(xy_data.PhoneService)
xy_data.MultipleLines = le.fit_transform(xy_data.MultipleLines)
xy_data.InternetService = le.fit_transform(xy_data.InternetService)
xy_data.OnlineSecurity = le.fit_transform(xy_data.OnlineSecurity)
xy_data.OnlineBackup = le.fit_transform(xy_data.OnlineBackup)
xy_data.DeviceProtection = le.fit_transform(xy_data.DeviceProtection)
xy_data.TechSupport = le.fit_transform(xy_data.TechSupport)
xy_data.StreamingTV = le.fit_transform(xy_data.StreamingTV)
xy_data.StreamingMovies = le.fit_transform(xy_data.StreamingMovies)
xy_data.Contract = le.fit_transform(xy_data.Contract)
xy_data.PaperlessBilling = le.fit_transform(xy_data.PaperlessBilling)
xy_data.PaymentMethod = le.fit_transform(xy_data.PaymentMethod)
xy_data.dropna(inplace=True)

xy_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

X = xy_data.drop('Churn', axis=1) #'Churn'열을 drop하고 input X를 정의함
classle=LabelEncoder() #class label
#'Churn'열의 문자열을 categorical value로 변환
y=classle.fit_transform(xy_data['Churn'].values) 
X.shape
```

```python
from sklearn.model_selection import train_test_split

#확인과정
X_train,X_valid,y_train,y_valid = train_test_split(X,y, test_size=0.3, random_state=1, stratify=y) 
print('입력변수(학습데이터)    : ',  X_train.shape)
print('입력변수(검증데이터)  : ', X_valid.shape)
print('종속변수(학습데이터)    : ',  y_train.shape)
print('종속변수(검증데이터)  : ', y_valid.shape)
```

```python
#KNN 적용
from sklearn.neighbors import KNeighborsClassifier  #KNN 불러오기
knn = KNeighborsClassifier(n_neighbors=8, p=1) 
knn.fit(X_train,y_train) 
y_train_pred = knn.predict(X_train) 
y_valid_pred  = knn.predict(X_valid)  
```

```python
# 정확도
from sklearn.metrics import accuracy_score 
print(accuracy_score(y_valid, y_valid_pred))
```

```python
from sklearn.metrics import confusion_matrix 
conf=confusion_matrix(y_true=y_valid, y_pred=y_valid_pred) 
print(conf)
```
