Likelihood(종속변수 y의 클래스 값에 따른 독립변수 X의 확률분포에 대한 정보)에 기반해서 베이즈 정리를 전개함.  
 


베이즈정리  
결합확룰은 곱셉법칙을 이용하여 두가지 방향으로 인수분해가 가능함.  
p(a, b) = p(b|a)p(a) = p(a|b)p(b)  
  
a를 관심의 대상인 종속변수에 해당될 때, 베이즈 정리는 p(b|a)를 이용하여 p(a|b)를 측정할 수 있도록 도와줌  
  
사전확률: 사전에 관측하기 전에 어떨 것이다 라는 것
사후확률: 사후에 관측 한 후에 이럴 것이다 라는 것.  
  

evidence는 한번에 알 수 없으므로 sum rule을 적용함.  
sum rule
b만 알기 어려움으로 b와 a가 같이 있는 경우 혹은 b에서 a를 뺀 값을 구함.
통계적추론의 2가지 방법  
MLE: Likelihood를 최대한 큰 값을 찾겠다.  
MAP: Likelihood에서 나오는 세타에 대한 확률분포를 찾겠다.
판별분석  
  
확률론적 생성모형 Likelihood를 기반해서 x의 확률분포에 대한 정보를 찾음.  
  
분류문제는 각 클래스 k에 대한 확률을 비교하여 가장 큰 값을 선택함.  
따라서 모든 클래스에 대해 값이 같은 분모P(x)는 계산 필요없음.   

이때 사전확률은 다음과 같이 계산한다.  
p(y=k) y=k인 데이터의 수 / 모든 데이터의 수   


Likelihood y에 대한 x의 조건부 확률  
특정한 확률분포모형을 따른다고 가정함.  
k번쨰 클래스에 속하는 학습데이터의 분포를 참조하여 결정함.  
판별분석에서는 이를 정규분포로 가정한다.
 

## 선형판별분석  LDA    
(Linear Discriminant Analysis)  

기본전제  
각 클래스 집단은 다변량 정규분포 형태의 확률분포를 가짐.  <------ 너무 하나가 크거나 작거나 하면 분류작업하기 이상해짐.      
각 클래스 집단은 비슷한 형태의 공분산 구조를 가짐.    
    2개의 확률변수의 상관정도를 나타내는 값.  
    공분산 행렬  
      
    공분산의 패턴  
    완전 공분산 상관관계가 있음. 
    대각 공분산  상관관계가 없음.
    구형 공분산   상관관계가 없음.    
      


### 선형 방정식이 됨.
판별함수가 X에 대한 선형함수가 됨.  <----- 계산하면 선형함수로 나옴.
  
선형판별분석의 기능      
2차원(2 개의 독립변수)의 두가지 범주(초록 파랑)을 갖는데 이것을 바꿈.  

LAD는 차원 축소의 개념을 포함함.  
2차우너 자료들을 판별 축에 정사영 시킨 분포의 형태를 고려함.  
  


### LAD 결정 경계의 특징  
projection 축에 직교하는 축  <------ 수선의 발을 내리는 형태 
각 클래스 집단의 평균의 차이가 크고 분산이 작은 지점을 결정 경계로 지정  
  
사영 데이터의 분포에서 겹치는 영역이 작은 결정 경계를 선텍.(분산 대비 평균의 차이를 극대화하는 결정경계를 찾음)  

# 실습#1 - 데이터 생성
- LDA

```python
from scipy.stats import multivariate_normal     # multivariate_normal 다변량 정규분포       # scipy 사이파이 수치해석을 위한 패키지 
import numpy as np

N = 100 #sample count
rv1 = multivariate_normal([ 0, 0], [[0.7, 0.0], [0.0, 0.7]]) #mean, cov     평균, 공분산 class1 위치(0, 1)  를 기준으로 분산
rv2 = multivariate_normal([ 1, 1], [[0.8, 0.2], [0.2, 0.8]]) #mean, cov     평균, 공분산 class2 위치(1, 1) 를 기준으로 분산
rv3 = multivariate_normal([-1, 1], [[0.8, 0.2], [0.2, 0.8]]) #mean, cov     평균, 공분산 class3 위치(-1, 1) 를 기준으로 분산

np.random.seed(0) #random seed      랜덤한 값이 나온걸 시드 0 에다 저장함. 그러면 랜덤하게 추출된 값이 0 시드에 저장됨 각 클래스마다 100개씩 넣겠다는 뜻 

X1 = rv1.rvs(N) #draw random samples from a multivariate normal distribution    다변량 정규본포에 랜덤 샘플을 그리겠다는 것.
X2 = rv2.rvs(N)
X3 = rv3.rvs(N)

y1 = np.zeros(N)  #class: 0     0으로 가득찬 클래스 y는 0
y2 = np.ones(N)   #class: 1     1로 가득찬 클래스 y는 1
y3 = 2*np.ones(N) #class: 2     2로 가득찬 클래스 y는 2
X = np.vstack([X1, X2, X3])     # 옆으로 stack
y = np.hstack([y1, y2, y3])     # 위로 stack 

print(X1.shape)
print(X2.shape)
print(X3.shape)

print(X1, "\n")
print(X2, "\n")
print(X3, "\n")
```

```python
import matplotlib.pyplot as plt

plt.scatter(X1[:,0], X1[:,1], alpha=0.8, s=50, marker="o", color='r', label="class 1") #alpha: marker의 투명도  # 첫번쨰 feature 두번쨰 feature
plt.scatter(X2[:,0], X2[:,1], alpha=0.8, s=50, marker="s", color='g', label="class 2")  
plt.scatter(X3[:,0], X3[:,1], alpha=0.8, s=50, marker="x", color='b', label="class 3")
plt.xlim(-5, 5)
plt.ylim(-4, 5)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.legend()
plt.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis        # 선형판별분석  LDA import
lda = LinearDiscriminantAnalysis(store_covariance=True).fit(X, y)       # X, y fiting       # store_covariance 공분산에 대한 결과 값을 찍어달라 
```

```python
lda.means_      # fiting 한 것 평균
```

```python
lda.covariance_     # fiting 한 것 분산     x1과 x2의 상관정도 완전고운산의 형태
```

```python
import seaborn as sns
import matplotlib as mpl

x1min, x1max = -5, 5
x2min, x2max = -4, 5
# meshgrid 격자 만들기 n 차원 직사각형 격자를 만드는 함수   # (x1max-x1min)/1000으로 나눠주고 arange를 적용함. 1000*1000 개의 시험 데이터가 생성이 됨. 
# meshgrid 함수 각각의 x, y 의 각각의 축들의 좌표를 입력받아, 2차원의 X, Y 좌표로 반환함.  
  
### xxx1 xx2에 x값 yy에 y 값  lda로 함
XX1, XX2 = np.meshgrid(np.arange(x1min, x1max, (x1max-x1min)/1000),     
                       np.arange(x2min, x2max, (x2max-x2min)/1000))
YY = np.reshape(lda.predict(np.array([XX1.ravel(), XX2.ravel()]).T), XX1.shape) 

cmap = mpl.colors.ListedColormap(sns.color_palette(["r", "g", "b"]).as_hex())       # 색상 지정     .as_hex() rgb에 핵사코드로 지정함. 
plt.contourf(XX1, XX2, YY, cmap=cmap, alpha=0.5)        # x와 y 값을 이용해서 동일한 값의 경계를 만들어주고, 경계에 따라 만들어진 각 그룹의 색상을 지정해줌.


plt.scatter(X1[:, 0], X1[:, 1], alpha=0.8, s=50, marker="o", color='r', label="class 1")           # x 범위 받아 적기
plt.scatter(X2[:, 0], X2[:, 1], alpha=0.8, s=50, marker="s", color='g', label="class 2")
plt.scatter(X3[:, 0], X3[:, 1], alpha=0.8, s=50, marker="x", color='b', label="class 3")
plt.xlim(x1min, x1max)
plt.ylim(x2min, x2max)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.legend()
plt.title("LDA Result")
plt.show()  


# 각 직선 경계(lda를 돌림) 색상 칠한거 부분은 클래스 1로 분류, 2로 분류, 3으로 분류 
```

# 실습#2 - Iris Data
- LDA

```python
#데이터 불러오기
import seaborn as sns #seaborn을 불러오고 네임스페이스를 sns로 축약
iris = sns.load_dataset('iris')  #iris라는 변수명으로 Iris 데이터셋 download

X = iris.drop('species', axis=1) #'species'열을 drop하고 input X를 정의
y_ = iris['species']
```

```python
#y data를 범주형으로 변환
from sklearn.preprocessing import LabelEncoder
classle = LabelEncoder() #class label
y = classle.fit_transform(iris['species'].values) #species 열의 문자열을 범주형으로 변환
```

```python
#전체 데이터를 학습과 테스트 데이터셋으로 분리
from sklearn.model_selection import train_test_split

#X(입력변수)와 y(종속변수)로 이루어진 데이터를 학습과 테스트 데이터넷으로 각각 70%, 30%의 비율로 나눔
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3, random_state=1, stratify=y) 
```

```python
X.head()
```

```python
y_.head()
```

```python
#Iris data에 대한 LDA 적합
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis        # 사이킷런에서 LDA 호출
lda=LinearDiscriminantAnalysis(store_covariance=True)   # 아무것도 없으니 사영축 개수는 디폴트 값인 2개가 출력됨.
#디폴트값: 
#lda=LinearDiscriminantAnalysis(n_components=2, store_covariance=True)  컴포넌트개수는 2 n_components  사영축 개수 디폴트값은 2개. 

lda.fit(X_train, y_train) # LDA 적합
y_train_pred=lda.predict(X_train)   #학습과 테스트 데이터를 가지고 예측함. 
y_test_pred=lda.predict(X_test)
```

```python
#Accuracy score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train,y_train_pred))  
print(accuracy_score(y_test,y_test_pred))    
```

```python
#분류 결과
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_test_pred))  #각 행은 setosa, versicolor, virginica
```

```python
print('Means:\n', lda.means_)
print('Xbar:\n', lda.xbar_)
print('Cov:\n', lda.covariance_)
print('Intercept:\n', lda.intercept_)
print('Coef:\n', lda.coef_)
```

```python
#Scatter plot
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
cld=LinearDiscriminantAnalysis()
X_lda = cld.fit_transform(X_train, y_train)     # fit을 하고 시각화를 위해서 fit_transform을 함.

# 캄포넌트 값을 지정하지 않으면 mean 값을 기준으로 사영축을 2개 만듦.
from matplotlib import pyplot as plt
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.scatter(X_lda[:,0],X_lda[:,1],c=y_train,cmap='rainbow',alpha=0.7,edgecolors='b');
```

판별분석의 이해  
두 개 이상의 모집단에서 추출된 표본들의 분포를 이용하여 어느 모집단에서 추출된 것인지를 결정해 줄수 있는 기준을 찾는 분석법  
  
선형 판별분석  
Linear  
  
이차판별분석  
Quadratic  
  
LR VS LDA  
LR: Discriminative  베이지한   
LDA: 다변량 정규분포, 공통 공분산 가정  <------ 데이터의 분포들을 항상 봐야함. 
데이터의 분포가 가정과 부합할 경우에는 LR보다 LDA의 성능이 우세 <----- 만약 공분산 주고가 같다면 괜춘 사용 가능.
데이터의 분포가 가정과 어긋날 경우에는 LDA보다 LR의 성능이 우세 <---- 결정 결계가 선형이거나, 공분산 구조가 같지 않거나, 정규분포도가 정형이 아니라면 사용하게 됨.
