판별분석이란  
두 개 이상의 모집단에서 추출된 표본들의 분포를 이용하여 이 표본들이 어느 모집단에서 추출된 것인지를 결정해 줄 수 있는 (좋은)기준을 찾는 분석법. 즉 어떤 모집단에서 잘 분류 해줄 수 있는 기준을 찾아주는 것.  


판별변수  
어떤 집단에 속하는지 판별하기 위한 변수로서 독립 변수 중 판별력이 높은 변수  

    선택하는 방법  
        두 독립변수 중 하나를 판별변수를 선택하기 그것과 상관관계가 작은 독립변수를 선택함으로서 효과적인 판별 함수(상관관계가 적지만 판별 값이 더 큰 것을 이용하는 것이 유리함.)


판별함수  
각 집단을 가장 잘 구분해 낼 수 있는 판별식

판별점수

표본의 크기

종속 변수의 집단(클래스)에서 최소 크기가 독립변수의 개수보다 커야 함. 판별력을 좌우하는 것은 가장 적은 집단의 표본 수 이기 떄문(그렇기 떄문에 너무 작은 클래스가 있다면 오버샘플링 하는 방법이던 조치를 취해야 함.)

베이즈정리  
결합확룰은 곱셉법칙을 이용하여 두가지 방향으로 인수분해가 가능함.  
p(a, b) = p(b|a)p(a) = p(a|b)p(b)  
  
a를 관심의 대상인 종속변수에 해당될 때, 베이즈 정리는 p(b|a)를 이용하여 p(a|b)를 측정할 수 있도록 도와줌  
  
사전확률: 사전에 관측하기 전에 어떨 것이다 라는 것
사후확률: 사후에 관측 한 후에 이럴 것이다 라는 것.  
  

evidence는 한번에 알 수 없으므로 sum rule을 적용함.  
sum rule
b만 알기 어려움으로 b와 a가 같이 있는 경우 혹은 b에서 a를 뺀 값을 구함.

통계적추론의 2가지 방법  
MLE: Likelihood를 최대한 큰 값을 찾겠다.  
MAP: Likelihood에서 나오는 세타에 대한 확률분포를 찾겠다.

분류문제는 각 클래스 k에 대한 확률을 비교하여 가장 큰 값을 선택함.

이때 사전확률은 다음과 같이 계산한다.  
p(y=k) y=k인 데이터의 수 / 모든 데이터의 수
