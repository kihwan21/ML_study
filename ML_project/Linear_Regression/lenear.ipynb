선형회귀란?  
종속변수와 한 갱 이상의 독립변수와의 관계를 선형적으로 모델링하는 접근방식  
<br> 종속변수 - dependent variable 독립변수 - independent variable  
선형모델의 회귀계수를 데이터로부터 추정하여 입력 값에 대한 연속형 종속변수 값을 예측함.  
  
선형회귀방식  
단순선형회귀 - 독립변수 한 개로 종속변수와의 관계를 선형적으로 모델링  
다중선형회귀 - 독립변수 2개 이상으로 종속변수와으이 관계를 선형적으로 모델링

모형 파라미터(회귀 계수)의 추정  
비용함수를 최소화하는 파라미터 세타0 세타1 의 값을 찾아내는 것.  
세타 데이터가 0, 1 두 개가 필요한 이유? 선형이 무조건 원점을 지난다는 보장이 없기 떄문에 편향(bias)를 줌.   


### 경사하강의 개념  세타 0 세타 1의 값을 어떻게 찾는가?


경사를 강하하면서 비용함수 J세타를 최소화하는 파라미터(세타)의 값을 찾아내는 방법  
  
경사하강 알고리즘 - 단순선형회귀  
  
목표 = argmin J(세타0, 세타1)  
  
작동방법: 파라미터값 초기화, 비용함수의 값이 최소가 될 떄까지 파리미터 값을 지속적으로 업데이트

비용함수의 (파라미터)값이 변하지 않을때까지   
  
경사하강의 learning rate(a)에 따라서 작동하는 시간이 다름. a가 작다면 수렴과정이 느리고 너무 많으면 발산하게 됨. 그렇기에 learning rate의 a가 중요함. 적절한 값으로 들어가있다면 결국 알맞은 값을 찾게됨.
