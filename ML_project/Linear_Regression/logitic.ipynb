로지스틱 함수는 S커브 함수를 나타냄.  
실제 많은 자연, 사회현상에서는 특정 변수에 대한 확률 값이 선형이 아닌 S 커브 형태를 따르는 경우가 많음  
X값으로 어떤 값이든 받을 수 있지만 출력결과 Y는 항상 0과 1 사이의 값을 내놓음.      

s커브를 적용하므로 몇이면 true, 몇 이하면 false로 나타내고싶다는 것.

로지스틱 회귀 가설함수의 해석. (확률)  
Estimated probability that y = 1 on input x          
x = [x0, x1] = [1/종양크기] 일때 h0(x) = 0.7 이라면  악성종양일 확률이 70%라고 해석할 수 있다.  
  
로지스틱 회귀 함수는 확률로서 표현할 수 있다.  
h0(x) = P(y = 1|x;0) probability that y = 1, given x, paramerterized by 0  
x가 주어졌을 때 y가 0이거나 1일 확률은 세타이다.  

암일때: 암일 확률 0.3 암이 아닐확률 0.7 = 1
